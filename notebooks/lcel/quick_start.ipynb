```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# This script initializes a ChatOpenAI model with the 'gpt-4o' model type.
model = ChatOpenAI(model='gpt-4o')

# It creates a chat prompt template with a system message and a user message.
# prompt = ChatPromptTemplate.from_template(
#     messages=[
#         ("system", "You are a helpful assistant."),
#         ("user", "Tell a joke about {input}."),
#     ]
# )
prompt = ChatPromptTemplate.from_template(template='Tell a joke about {input}.')

# It sets up an output parser to parse the model's response as a string.
output_parser = StrOutputParser()

# The script then combines the model, prompt, and output parser into a single chain.
chain = prompt | model | output_parser

# Finally, it runs the chain with a specific input and prints the output.
output = chain.invoke({"input": "dogs"})
print(output)
# The following line streams the output from the chain, allowing for real-time output display.
# for s in chain.stream({"input": "dogs"}):
#     print(s, end='', flush=True)

```
